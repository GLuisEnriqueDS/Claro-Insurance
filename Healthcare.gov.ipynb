```python
carrier_list = []
plan_name_list = []
monthly_premium_list = []
tax_credit_list = []
was_tax_credit_list = []
group_information_list = []
deductible_list = []
out_of_pocket_list = []
calification_list = []
benefits_list = []
zipcode_list = []
county_list =[]
income_list =[]

zipcodes = ['33012','33024','33411','33647','34787','32218','33810','32174','32765','34221','34711','34744','34953','32608','32259','32068','32162','30349','30044','30058','30127','31419','30188','30281','30236','30506','30906','31907','30040','31210','30809','31088','30135','30263','30180','30606','30052','31602','30269','30157','30016','30721','30120','30165','31705','30094','30680','31525','30458','30240','31313','30223','30701','30736','31324','31061','30705','27610','28269','27406','27284','27703','28314','28806','28173','28412','28027','28540','28117','27215','27858','27520','28601','27516','27360','28092','28306','28655','28645','27330','28376','27909','29681','29223','29464','29588','29072','29732','29485','29910','29621','29445','29501','29720','38017','37013','37918','37421','37128','37064','37042','37075','37660','37604','37122','37803','37312','37876','38305','38501','38401','37830','37172','37814','37355','37160','37643','37110','77449','75052','76063','78245','78660','75035','75067','79936','78542','77494','78613','77386','78521','77584','76549','77573','78414','79424','78046','76706','77705','78666','75703','77845','75165','76028','78130','79762','79705','79605','75605','78155','75092','79109','75126','76087','79107','76904','75087','78602','75501','77901','77535','77630','75904','77320','76522','78582','76049','78852','78028','75110','77351','78840','78006','76401','78332','77523','76240','77833','79720','78363','75455','78102']
income_value = 25000
age_value = 25

li_ids = ["autocomplete--9__item--0", "autocomplete--9__item--1", "autocomplete--9__item--2"]


for zip in zipcodes:
    for li_id in li_ids:
        try:
            opts = Options()
            driver = webdriver.Chrome('./chromedriver.exe' ,chrome_options=opts)
            driver.delete_all_cookies()
            driver.get('https://www.healthcare.gov/see-plans/#/?year=2023')
            driver.maximize_window()
            #Enter your ZIP Code & choose your location:
            choose = WebDriverWait(driver, 60).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='ZIP Code']")))
            choose.clear()
            choose.send_keys(str(zip))
            # Seleccionamos locaci√≥n
            plan = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, f'//li[@id="{li_id}"]')))
            plan.click()
            input = driver.find_element(By.XPATH,'//input[@class="ds-c-field"]')
            county_list.append([zip,input.get_property("value")])
            sleep(random.uniform(2.0, 3.0))  
            # Localizamos botom Continuar  
            button = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//button[contains(@class,"ds-c-button ds-c-button--solid ds-u-sm-margin-left--1 ds-u-margin-top--05")]')))
            button.click() 
            sleep(random.uniform(2.0, 3.0))  
            # Tell us about your household
            button = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//a[contains(@class,"ds-c-button ds-c-button--solid ds-c-button--main ds-c-button--on-light")]')))
            button.click() 
            # Are you enrolled in a 2022 Marketplace health plan?
            button = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//button[contains(@class,"ds-c-button ds-c-button ds-c-button--secondary qa-skip")]')))
            button.click()  
            # Who is in your household: Just You
            input = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//input[contains(@class,"ds-c-choice")][1]')))
            input.click() 
            # Who is in your household: Continue
            button = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//button[contains(@class,"ds-c-button ds-c-button--solid")]')))
            button.click() 
            # Tell us about you: Sex
            age = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//input[@class="ds-c-field ds-c-field--small"]')))
            age.click() 
            age.clear() 
            age.send_keys(str(age_value))
            # Tell us about you: Gender
            gender = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//input[contains(@class,"ds-c-choice")][1]')))
            gender.click()
            # Tell us about you: Continue
            sleep(3)
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            sleep(3)
            button = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//button[contains(@class,"ds-c-button ds-c-button--solid qa-continue")]')))
            button.click()
            
            # Confirm your household members
            button = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//button[contains(@class,"ds-c-button ds-c-button--solid qa-continue ds-u-display--block ds-u-margin-y--2")]')))
            button.click() 
            # What do you think your household income will be in 2023?
            household_income= WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH,'//input[@class="ds-c-field ds-c-field--medium ds-c-field--currency"]')))
            household_income.click() 
            household_income.clear() 
            household_income.send_keys(str(income_value))
            # What do you think your household income will be in 2023? Continue
            button = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//button[contains(@class,"ds-c-button ds-c-button--solid qa-continue-btn")]')))
            button.click() 
            # View Plans
            try:
                button = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//button[contains(@class,"ds-c-button ds-c-button--solid qa-continue-btn")]')))
                button.click()   
            except:
                button = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//button[contains(@class,"ds-c-button ds-c-button--ghost qa-continue-link ds-c-button ds-u-padding--0 ds-u-font-weight--normal")]')))
                button.click() 
            # Localizamos boton "Close"
            button = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,'//button[@class="ds-c-button ds-c-button--ghost ds-c-dialog__close"]')))
            button.click()
            sleep(random.uniform(2.0, 3.0))  

            while True:
                WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//article[@class="pet-c-plan-card--new ds-base ds-u-fill--white"]')))

                casillas = driver.find_elements(By.XPATH, '//li[@class="ds-u-margin-bottom--4"]')

                for casilla in casillas:

                    driver.execute_script("arguments[0].scrollIntoView(true);", casilla)
                    sleep(0.5)

                    # CARRIER
                    try:
                        carrier = casilla.find_element(By.XPATH,'.//div[@class="pet-c-plan-title__issuer ds-u-font-weight--bold ds-u-font-size--sm"]').text
                    except:
                        carrier = 'Error'
                    carrier_list.append(carrier)
                    # PLAN NAME
                    try:
                        plan_name = casilla.find_element(By.XPATH,'.//h2[@class="pet-c-plan-title__name ds-u-font-weight--normal ds-u-margin-y--1 ds-u-font-size--lg"]').text
                    except:
                        plan_name = 'Error'
                    plan_name_list.append(plan_name)
                    # ESTIMATED MONTHLY PREMIUM
                    try:
                        monthly_premium = casilla.find_element(By.XPATH,'.//div[@class="pet-c-plan__currency-container ds-u-font-size--base ds-u-md-font-size--3xl"]').text
                    except:
                        monthly_premium = 'Error'
                    monthly_premium_list.append(monthly_premium)  
                    # TAX CREDIT
                    try:
                        tax_credit = casilla.find_element(By.XPATH,'.//div[@class="pet-c-plan-cost__explanation ds-u-font-size--small ds-u-color--muted"][1]').text
                    except:
                        tax_credit = 'Error'
                    tax_credit_list.append(tax_credit)  
                    # TAX CREDIT WAS
                    try:
                        was_tax_credit = casilla.find_element(By.XPATH,'.//div[@class="pet-c-plan-cost__premium-without-credit ds-u-font-size--small ds-u-font-style--italic ds-u-color--muted"]').text
                    except:
                        was_tax_credit = 'Error'
                    was_tax_credit_list.append(was_tax_credit)  
                    # GROUP INFORMATION
                    try:
                        group_information = casilla.find_element(By.XPATH,'.//div[@class="ds-u-display--flex ds-u-align-items--center ds-u-flex-wrap--wrap pet-c-plan-title-group pet-c-plan-title-group3"]').text
                    except:
                        group_information = 'Error'
                    group_information_list.append(group_information)
                    # DEDUCTIBLE
                    try:
                        deductible = casilla.find_element(By.XPATH,'.//div[@class="pet-c-plan__currency-container ds-u-font-size--base ds-u-md-font-size--2xl ds-u-margin-top--1 ds-u-md-margin-bottom--1"]').text
                    except:
                        deductible = 'Error'
                    deductible_list.append(deductible)  
                    # OUT OF POCKET
                    try:
                        out_of_pocket = casilla.find_element(By.XPATH,'.//div[@class="ds-c-cost-display__amount pet-c-plan__currency-container"]/div[@class="pet-c-plan__currency-container ds-u-font-size--base ds-u-md-font-size--2xl ds-u-margin-top--1 ds-u-md-margin-bottom--1"]').text
                    except:
                        out_of_pocket = 'Error'
                    out_of_pocket_list.append(out_of_pocket)  

                    # BENEFITS
                    try:
                        WebDriverWait(casilla,15).until(EC.element_to_be_clickable((By.XPATH,'.//button[text()="Check what you pay when you get care"]')))
                        button = casilla.find_element(By.XPATH,'.//button[text()="Check what you pay when you get care"]')
                        button.click()
                        benefits_div = casilla.find_element(By.XPATH, './/div[contains(@class, "pet-c-plan-card--benefits")]')
                        benefits = benefits_div.text
                    
                    except Exception as e:
                        benefits = 'Error'
                    
                    benefits_list.append(benefits) 
                    
                    # CALIFICATION
                    try:
                        calification = casilla.find_element(By.XPATH,'.//div[@class="ds-u-align-items--center pet-c-plan-title__info-item pet-c-plan-title__info-item__ratings ds-u-font-size--sm ds-u-display--flex"]').text
                    except:
                        calification = 'Error'
                    calification_list.append(calification) 
                    # ZIP CODE
                    zipcode_list.append(str(zip))
                    # INCOME
                    income_list.append(str(income_value))
                try:
                    driver.execute_script("window.scrollBy(0, document.body.scrollHeight);")
                    next_link = WebDriverWait(driver,15).until(EC.element_to_be_clickable((By.XPATH,'//a[@class="ds-c-button ds-c-button--ghost ds-c-pagination__nav" and  @aria-label="Next Page"]')))
                    next_link.click() 

                except:
                    break
        except:
            pass
            
simple_list = pd.DataFrame(county_list, columns=["zip code","County"])     
informacion_planes = pd.DataFrame({'carrier':carrier_list,
                                'plan_name':plan_name_list,
                                'monthly_premium':monthly_premium_list,
                                'tax_credit':tax_credit_list,
                                'was':was_tax_credit_list,
                                'group_information':group_information_list,
                                'deductible':deductible_list,
                                'out_of_pocket':out_of_pocket_list,
                                'benefits':benefits_list,
                                'calification':calification_list,
                                'zip code':zipcode_list,
                                'income':income_list}).merge(simple_list,left_on='zip code', right_on= 'zip code', how = 'left')


informacion_planes
```

```python

df = informacion_planes.drop_duplicates()
df = df[df['carrier'] != 'Error']
df['monthly_premium'] = df['monthly_premium'].replace('/month', '', regex=True).replace('\$', '', regex=True).replace(',', '', regex=True).astype(float)
df['tax_credit'] = df['tax_credit'].str.split('\n').str[0]
df = df.rename(columns={'was': 'premium_without_credit',
                        'County':'location',
                        'tax_credit':'premium_explanation',
                        'monthly_premium':'premium',
                        'plan_name':'plan'})

df[['metal', 'plan_type', 'plan_id']] = df['group_information'].str.extract(
    'Metal Level:\n(.+)\nPlan type:\n(.+)\nPlan ID:\n(.+)', 
    expand=True
)

def extract_benefit_info(benefit_str, benefit_type):
    import re
    
    patterns = {
        "Primary care": r"Primary care\n(.+?)\n",
        "Specialist care": r"Specialist care\n(.+?)\n",
        "Urgent care": r"Urgent care\n(.+?)\n",
        "Emergency room": r"Emergency room\n(.+?)\n",
        "Outpatient mental health": r"Outpatient mental health\n(.+?)\n",
        "Generic drugs": r"Generic drugs\n(.+?)\n",
    }
    
    match = re.search(patterns[benefit_type], benefit_str)
    return match.group(1) if match else None

benefit_types = ["Primary care", "Specialist care", "Urgent care", "Emergency room", "Outpatient mental health", "Generic drugs"]
for benefit_type in benefit_types:
    df[benefit_type] = df['benefits'].apply(lambda x: extract_benefit_info(x, benefit_type))

df['calification'] = df['calification'].replace(r'Rating\\n', '', regex=True)

df[['location','carrier','plan','metal','plan_type','plan_id','premium','premium_explanation','premium_without_credit','deductible','out_of_pocket','Primary care','Specialist care','Urgent care','Emergency room','Outpatient mental health','Generic drugs','income','zip code','benefits']].to_excel('ejemplo resultado final.xlsx')
```
